```python
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >>i) for j in range(N)]
        total += sum(L)
    return total
```

```python
%%file mprun_demo.py
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
        del L # remove reference to L
    return total
```
